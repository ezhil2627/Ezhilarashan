# ** FINAL CONCLUSION**

**Based on the visual comparison of the four confusion matrices (CM1, CM2, CM3, CM4), CM4 (Standardization) appears to be the best-performing model.**

**Reasoning**:

**CM4**: **Standardization shows the darkest and most concentrated color along the diagonal of its heatmap. This indicates that it had the highest percentage of correct predictions (true positives and true negatives).**



**The off-diagonal elements in CM4 are generally lighter, indicating fewer incorrect predictions compared to the other models.**



**While CM2 (Outlier Removal) and CM3 (L1 Normalization) might show some improvement over the baseline CM1, they do not exhibit the same level of accuracy as CM4.**



**Therefore, applying standardization as a data preprocessing technique (as seen in CM4) led to the most accurate KNN model for this dataset.**
